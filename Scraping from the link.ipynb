```python
import os
from selenium import webdriver
# Import sleep:
from time import sleep
import pandas as pd
from bs4 import BeautifulSoup
import numpy as np
import re

chromedriver = "./chromedriver"
os.environ["webdriver.chrome.driver"] = chromedriver

driver = webdriver.Chrome(executable_path="./chromedriver")
# driver = webdriver.Chrome(chromedriver)
```

```python
df= pd.read_csv('df_1Feb.csv')

links=df.links
```

```python
links_800=links[200:839]
#links_800
```

```python
df.shape
```

```python
ref=[]
job_titles=[]
companys=[]    
employment_types=[]
gov_support=[]
industrys=[]
salary=[]
salary_terms=[]
job_descriptions=[]
requirements=[]
senioritys=[]


for link in links_800:
   
    print(link)
    ref.append(link)
    driver.get(link)
    sleep(10)
    # Grab the page source.div class="bg-white pa4"
    html = driver.page_source
    # Beautiful Soup it!
    html = BeautifulSoup(html, 'lxml')
    for entry in html.find_all('div', {'class':'bg-white pa4'}):
  
        #Grab the Jobtitle.
       
        job_title = entry.find('h1', {'class': 'f3 fw6 mv0 pv0 mb1 dark-pink w-100 dib'}).text
        job_titles.append(job_title)
        #print(job_title)
    
        # Grab the Company.
        try:
            company = entry.find('p', {'class': 'f6 fw6 mv0 black-80 mr2 di ttu'}).text
            companys.append(company)
        except:
            companys.append(np.nan)
            
        #print(company)
    
        #seniorty
        try:
            seniority=entry.find('p',{'class': 'black-60 f6 fw4 mv1 dib pr3 mr1 icon-bw-seniority'}).text
            senioritys.append(seniority)
        except:
            senioritys.append(np.nan)
        
        #print(seniority)
    
        #employment_type
        try:
            employment_type=entry.find('p',{'class':'black-60 f6 fw4 mv1 dib pr3 mr1 icon-bw-employment-type'}).text
            employment_types.append(employment_type)
        except:
            employment_types.append(np.nan)
        
        #print(employment_type)
    
        # gov_support
        
        try:
            gov_supp=entry.find('p',{'class':'f5 fw7 pa0 mh0 mt0 mb4 brand-sec'}).text
            gov_support.append(gov_supp)
        except:
            gov_support.append(np.nan)
        
        #industry:
        try:
            industry=entry.find('p',{'class':'black-60 f6 fw4 mv1 dib pr3 mr1 icon-bw-category'}).text
            industrys.append(industry)
        except:
            industrys.append(np.nan)
        
        #print(industry)
    
        #salary:
    
        for row in entry.findAll("span", class_ = re.compile("salary_range")):
            salary.append(row.text)
            
            #print(salary)
            
        #monthly/yearly
        
        try:
            salary_term=entry.find('span', {'class':'salary_type dib f5 fw4 black-60 pr1 i pb'}).text
            salary_terms.append(salary_term)
        except:
            salary_terms.append(np.nan)
            
        #print(salary_term)
            
        #job description
        try:
            job_description=entry.find('div', {'id': 'description-content'}).text
            job_descriptions.append(job_description)
        except:
            job_descriptions.append(np.nan)
            
        #print(job_description)
    
    #requirements
        try:
            requirement=entry.find('div', {'id': 'requirements-content'}).text
            requirements.append(requirement)
        except:
            requirements.append(np.nan)
            
        #print(requirement)
            

# First, create an empty DataFrame.
import pandas as pd


column_headers=["job_titles","companys", "employment_type","seniority","gov_support", 
         "industrys","salary","salary_terms","job_descriptions","job_requirements", "links"]

jobs21 = pd.DataFrame(columns= column_headers)

jobs21["job_titles"]=job_titles

jobs21["companys"]=companys

jobs21["employment_type"]=employment_types

jobs21["seniority"]= senioritys

jobs21["gov_support"]=gov_support

jobs21["industrys"]=industrys

jobs21["salary"]=salary

jobs21["salary_terms"]=salary_terms

jobs21["job_descriptions"]=job_descriptions

jobs21["job_requirements"]=requirements

jobs21["links"]=ref


jobs21.to_csv('jobs_1Feb_800.csv')

driver.close()
     
```
